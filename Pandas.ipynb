### **CSV**

```python
from io import StringIO, BytesIO
```

```python
data = ('col1,col2,col3\n'
        'x,a,1\n'
        'y,b,2\n'
        'z,c,3')
```

```python
type(data)
```

```python
StringIO
```

```python
pd.read_csv(StringIO(data))
```

```python
df=pd.read_csv(StringIO(data),usecols=['col1','col2'])
```

```python
df
```

```python
### Specifying columns data types
data = ('a,b,c,d,e\n'
       '1,2,3,4,5\n'
       '5,6,7,8,9\n'
       '9,10,11,12,13')
```

```python
print(data)
```

```python
df = pd.read_csv(StringIO(data),dtype=object)
```

```python
df
```

```python
df['a'][2]
```

```python
df = pd.read_csv(StringIO(data),dtype=int)
df
```

```python
df['c']
```

```python
df = pd.read_csv(StringIO(data),dtype=float)
df
```

```python
df=pd.read_csv(StringIO(data),dtype={'b':int,'e':float,'d':object,'c':'int64'})
df
```

```python
type(df['e'][1])
```

```python
df.dtypes
```

```python
#### Index Columns and training delimeters
data = ('index,a,b,c,d\n'
       '1,apple,bat,5.7,Male\n'
       '2,orange,cow,10,Female')
```

```python
pd.read_csv(StringIO(data))
```

```python
pd.read_csv(StringIO(data),index_col=0)
```

```python
data = ('a,b,c\n'
       '1,apple,bat,\n'
       '2,orange,cow,')
```

```python
pd.read_csv(StringIO(data))
```

```python
pd.read_csv(StringIO(data),index_col=False)
```

```python
## Combining usecols & index_col
pd.read_csv(StringIO(data),usecols=['a','c'],index_col=False)
```

```python
### Quoting & Escape characters
data ='a,b\n"Hello, \\"Sony\\", nice to see ya",7'
```

```python
pd.read_csv(StringIO(data),escapechar='\\')
```
